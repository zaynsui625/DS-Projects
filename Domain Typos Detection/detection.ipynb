### Define valid domains

```python
pop = pd.read_csv(r"C:\Ziyuan Sui\RazorSQL\Validation\Email Validation\domain_popularity.csv")
```

```python
pop.head(14)
```

```python
pop.dropna(inplace=True)
```

```python
valid_domains= []
for index,row in pop.iterrows():
    if row['domain_popularity']>4:
        valid_domains.append(row['domain'])
print(valid_domains)
```

### Functions: detect domain typos

```python
similarity_scores = []
decisions=[]
threshold = 80
for index, row in df.iterrows():
    domain = row['domain']
    max_similarity_score=-1

    for valid in valid_domains:
        similarity_score = fuzz.ratio(domain, valid)
        max_similarity_score = max(max_similarity_score,similarity_score)
    
    if max_similarity_score < threshold:
        decision = "Need to double check manually"
    elif max_similarity_score == 100:
        decision = "Valid"
    else:
        decision = "Typo"
    
    similarity_scores.append(max_similarity_score)
    decisions.append(decision)


df['similarity_score'] = similarity_scores
df['decision'] = decisions
```
